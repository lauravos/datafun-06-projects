## Section 7-List: 
### Get some of your information into a list. Process each item in the list (use for or comprehensions as you like). 

```python
#turned weights into a list
weightlist = cats['weight'].tolist()
weightlist
```

```python
#turned breeds into list
breedslist = cats['breed'].tolist()
breedslist
```

## Section 8-Filter: 
### Use filter() to show only part of the information. 

```python
#filtered list of weights to only include weights over 20 lbs
weights_over_20 = list(filter(lambda x: x > 20, weightlist))
weights_over_20
```

```python
#used comprehension method as alternative way to filter list of weights to only include weights over 20 lbs as well
weights_over_20 = [x for x in weightlist if x > 20]
weights_over_20
```

## Section 9-Map: 
### Use map() to transform some of the data. 

```python
#use map to convert weight (in lbs) to kgs (conversion factor of 0.453592kg/1lb)
weight_in_kgs = list(map(lambda x: x * 0.453592, weightlist))
weight_in_kgs
```
